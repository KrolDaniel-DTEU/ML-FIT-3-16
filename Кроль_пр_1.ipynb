```python
# Варіант 8
# Кроль 3-16

import numpy as np
```

```python
# Завдання 1

# Створіть одновимірний масив (вектор) зі спадаючою послідовністю чисел, де кожне
# наступне число менше попереднього на 2, починаючи з 30, та виведіть його значення.

arr = np.arange(30, 0, -2)
print(arr)
```

```python
# Завдання 2

# Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його значеннями
# факторіалів натуральних чисел, починаючи з 1, та виведіть його значення.

numbers = np.arange(1, 10)
factorials = np.cumprod(numbers)
matrix = factorials.reshape(3, 3)

print(matrix)
```

```python
# Завдання 3

# Створіть масив розміром 4x4, заповніть його випадковими цілими числами
# від 1 до 10 та виведіть його значення

matrix = np.random.randint(1, 11, (4, 4))

print(matrix)
```

```python
# Завдання 4

# Створіть масив розміром 6x6, заповніть його випадковими дійсними числами в
# діапазоні від 0 до 1 та обчисліть суму всіх діагональних елементів

matrix = np.random.uniform(0, 1, (6, 6))
diagonal = np.diag(matrix)
sd = np.sum(diagonal)
print(f"{matrix=}")
print(f"{diagonal=}")
print(f"{sd=}")
```

```python
# Завдання 5

# Створіть два одновимірних масиви розміром 4, заповніть їх випадковими дійсними
# числами від -10 до 10 та виконайте на них поелементні операції додавання,
# віднімання та множення

arr1 = np.random.uniform(-10, 10, 4)
arr2 = np.random.uniform(-10, 10, 4)

print(f"{arr1=}")
print(f"{arr2=}")
print(f"{arr1 + arr2=}")
print(f"{arr1 - arr2=}")
print(f"{arr1 * arr2=}")
```

```python
# Завдання 6

# Створіть дві матриці 4x4, заповніть їх довільними цілими числами від 1 до 5 та
# знайдіть їх добуток

m1 = np.random.randint(1, 6, (4, 4))
m2 = np.random.randint(1, 6, (4, 4))

prod = np.dot(m1, m2)

print(f"{m1=}")
print(f"{m2=}")
print(f"{prod=}")
```

```python
# Завдання 7

# Створіть дві матриці розміром 2x3 та 3x3, заповніть їх випадковими
# цілими числами в діапазоні від 1 до 4 та перемножте їх між собою

m1 = np.random.randint(1, 5, (2, 3))
m2 = np.random.randint(1, 5, (3, 3))

prod = np.dot(m1, m2)

print(f"{m1=}")
print(f"{m2=}")
print(f"{prod=}")
```

```python
# Завдання 8

# Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в
# діапазоні від 1 до 7 та знайдіть її обернену матрицю

matrix = np.random.randint(1, 8, (3, 3))
invert = np.linalg.inv(matrix)

print(f"{matrix=}")
print(f"{invert=}")
```

```python
# Завдання 9

# Створіть матрицю розміром 5x5, заповніть її випадковими дійсними числами в
# діапазоні від -5 до 5 та транспонуйте її

matrix = np.random.uniform(-5, 5, (5, 5))
trans = matrix.transpose();

print(f"{matrix=}")
print(f"{trans=}")
```

```python
# Завдання 10

# Створіть матрицю розміром 3x4 та вектор розміром 4x1, заповніть матрицю
# випадковими цілими числами в діапазоні від 1 до 10, а вектор - випадковими дійсними
# числами від 0 до 1 і знайдіть добуток матриці на вектор

matrix = np.random.randint(1, 11, (3, 4))
arr = np.random.uniform(0, 1, 4)
prod = np.dot(matrix, arr)

print(f"{matrix=}")
print(f"{arr=}")
print(f"{prod=}")
```

```python
# Завдання 11

# Створіть матрицю розміром 2x3 та вектор розміром 3x1, заповніть матрицю
# випадковими дійсними числами від 0 до 1, а вектор - випадковими дійсними числами
# від 0 до 1, і знайдіть добуток матриці на вектор

matrix = np.random.uniform(0, 1, (2, 3))
arr = np.random.uniform(0, 1, 3)
prod = np.dot(matrix, arr)

print(f"{matrix=}")
print(f"{arr=}")
print(f"{prod=}")
```

```python
# Завдання 12

# Створіть дві матриці розміром 9x9, заповніть їх випадковими цілими числами в
# діапазоні від 1 до 10 та знайдіть відношення елементів другої матриці до першої
# (поелементне ділення)

m1 = np.random.randint(1, 11, (9, 9))
m2 = np.random.randint(1, 11, (9, 9))

print(f"{m1=}")
print(f"{m2=}")
print(f"{m1/m2=}")
```

```python
# Завдання 13

# Створіть дві матриці розміром 9x9, заповніть їх випадковими цілими числами в
# діапазоні від 1 до 5 та знайдіть їх добуток

m1 = np.random.randint(1, 6, (9, 9))
m2 = np.random.randint(1, 6, (9, 9))
prod = np.dot(m1, m2)

print(f"{m1=}")
print(f"{m2=}")
print(f"{prod=}")
```

```python
# Завдання 14

# Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні
# від -5 до 5 та знайдіть середнє арифметичне значення елементів матриці.

matrix = np.random.uniform(-5, 5, (3, 3))
mean = np.mean(matrix)

print(f"{matrix=}")
print(f"{mean=}")
```

```python
# Завдання 15

# Створіть матрицю розміром 10x10, заповніть її випадковими дійсними числами в
# діапазоні від -2 до 2 та знайдіть вектор-стовпчик, що містить найменше за модулем
# значення елементу у кожному рядку матриці.

matrix = np.random.uniform(-2, 2, (10, 10))
min_indices = np.argmin(np.abs(matrix), axis=1)
min_elements = matrix[np.arange(matrix.shape[0]), min_indices]
min_column_vector = min_elements.reshape(-1, 1)

print(f"{matrix=}")
print(f"{min_column_vector=}")
```
